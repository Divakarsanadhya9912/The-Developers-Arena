# Supermarket Sales Analysis

## Objective
To analyze supermarket sales data and extract business insights

```python
import pandas as pd
import sys
import os

sys.path.append(os.getcwd())

from cleaning import load_and_clean_data
from visualization import (
    plot_daily_sales,
    plot_product_line_sales,
    plot_branch_sales,
    plot_hourly_sales,
    plot_correlation_heatmap
)
```

```python
file_path = "../datasets/supermarket_sales.csv"

df = load_and_clean_data(file_path)

plot_daily_sales(df)
plot_product_line_sales(df)
plot_branch_sales(df)
plot_hourly_sales(df)
plot_correlation_heatmap(df)
```

```python
# -----------------------------
# BUSINESS KPIs
# -----------------------------

total_revenue = df["total"].sum()
total_transactions = df.shape[0]
average_transaction_value = df["total"].mean()
total_quantity_sold = df["quantity"].sum()
average_rating = df["rating"].mean()

print("Total Revenue:", round(total_revenue, 2))
print("Total Transactions:", total_transactions)
print("Average Transaction Value:", round(average_transaction_value, 2))
print("Total Quantity Sold:", total_quantity_sold)
print("Average Customer Rating:", round(average_rating, 2))
```

```python
top_branch = df.groupby("branch")["total"].sum().sort_values(ascending=False)
print("Top Performing Branch:")
print(top_branch)
```

```python
top_product = df.groupby("product_line")["total"].sum().sort_values(ascending=False)
print("Top Product Line:")
print(top_product)
```

```python
peak_hour = df.groupby("hour")["total"].sum().sort_values(ascending=False)
print("Peak Sales Hour:")
print(peak_hour.head())
```

## Executive Summary

The supermarket generated a total revenue of ₹519,280.69 from 2,000 transactions, 
with an average transaction value of ₹259.64. A total of 10,066 units were sold 
with an average customer rating of 6.96, indicating moderate customer satisfaction.


## Branch Performance Analysis

- Branch B is the top performing branch with ₹180,821.79 in revenue.
- Branch A follows closely with ₹179,047.93.
- Branch C generates the lowest revenue among the three branches.

📌 Insight:
Revenue distribution across branches is relatively balanced, 
but Branch B shows slightly stronger operational performance.


## Product Line Performance

Top revenue-generating product lines:

1. Health & Beauty – ₹96,970.32
2. Sports & Travel – ₹91,177.89
3. Electronic Accessories – ₹88,995.19

📌 Insight:
Health & Beauty is the strongest revenue driver, indicating high customer demand.
Food & Beverages and Fashion Accessories generate comparatively lower revenue.


## Customer Purchase Behavior

- Peak sales occur at 18:00 hours, followed by 14:00 and 08:00.
- Evening hours (around 6 PM) show the highest customer activity.

📌 Insight:
The supermarket experiences strong evening footfall. 
Marketing campaigns and staffing can be optimized around peak hours.


## Overall Business Observations

- Average transaction value of ₹259.64 suggests moderate basket size.
- Total quantity sold (10,066 units) reflects consistent product movement.
- Customer rating of 6.96 indicates room for service or product quality improvement.


## Strategic Recommendations

1. Focus inventory expansion on Health & Beauty products.
2. Increase marketing efforts during evening peak hours.
3. Investigate service improvements to increase customer rating.
4. Analyze why Branch C underperforms compared to A and B.
