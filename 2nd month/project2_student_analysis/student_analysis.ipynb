```python
import pandas as pd
from cleaning import load_data, basic_inspection, clean_data
from visualization import pass_fail_chart, subject_performance, attendance_correlation
```

```python
df = load_data("../datasets/student_performance_week2.csv")
basic_inspection(df)
```

```python
df = clean_data(df)
df.head()
```

```python
pass_fail_counts = df["pass_status"].value_counts()
pass_percentage = (pass_fail_counts / len(df)) * 100

print("Pass/Fail Counts:\n", pass_fail_counts)
print("\nPass/Fail Percentage:\n", pass_percentage)
```

```python
subject_avg = df[["math_score", "science_score", "english_score"]].mean()
print("Average Scores per Subject:\n", subject_avg)
```

```python
correlation = df["attendance_percentage"].corr(df["average_score"])
print("Attendance vs Average Score Correlation:", correlation)
```
